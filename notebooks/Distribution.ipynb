```python
import pandas as pd
df = pd.read_csv(r'C:\Users\camer\Downloads\WeatherV4.csv', sep='@')
df['datetime'] = pd.to_datetime(df['SampledDate'] + ' ' + df['SampledTime'])
df_sorted = df.sort_values(by='datetime')
print(df_sorted[0:10])
```

```python
count_per_datetime = df_sorted.groupby('datetime').size()
print(count_per_datetime)
```

```python
import matplotlib.pyplot as plt

plt.hist(count_per_datetime)
plt.xlabel('Count of Occurrences')
plt.ylabel('Frequency')
plt.title('Distribution of Occurrences per Datetime')
plt.show()

```

```python
frequency_count = count_per_datetime.value_counts().sort_index()
for count, frequency in frequency_count.items():
    print(f"{count}: {frequency}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter
year_counter = Counter()
with open(r'C:\Users\camer\Downloads\WeatherV4.csv', 'r') as f:
    next(f)
    for line in f:
        year = int(line.split('@')[1].split('-')[0])
        year_counter[year] += 1
total_relevant_years = sum(year_counter.values())
probability_distribution = {year: count / total_relevant_years for year, count in year_counter.items()}
plt.bar(probability_distribution.keys(), probability_distribution.values())
plt.xlabel('Year')
plt.ylabel('Probability')
plt.title('Probability Distribution of Years')
plt.show()
```
